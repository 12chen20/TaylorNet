# TaylorNet

```python
import pandas as pd
import math
import matplotlib.pyplot as plt
data= pd.read_csv("C:\\Users\\可乐怪\\Desktop\\data.csv")
data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from collections import namedtuple

n_groups = 3
means_men = data['y1'].iloc[:3]
means_women = data['y2'].iloc[:3]

 
fig, ax = plt.subplots()
 
index = np.arange(n_groups)
bar_width = 0.25
 
 
# rects1 = ax.bar(index, means_men, bar_width,
#                color='silver')
rects1 = ax.bar(index, means_men, bar_width,
               color='silver')
# # rects2 = ax.bar(index + bar_width, means_women, bar_width,
#                 color='darksalmon')
rects2 = ax.bar(index + bar_width, means_women, bar_width,
                color='darksalmon')

ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(data['x'].iloc[:3])
ax.legend()
plt.legend(['k=3','k=10'],loc=(1.1, 0.75),frameon=False,fontsize = 15)
ax.set_ylabel('Accuracy (%)',fontsize=20,fontproperties='Times New Roman')
plt.xticks( fontproperties='Times New Roman',fontsize = 15)
plt.ylim(84.5,90.2)
plt.savefig('2.pdf', dpi=1200, bbox_inches = 'tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.special import comb
# 定义函数 h(λ)
def h(lambda_1):
    theta = [1.0968, 0.9154, 1.0834, 0.918, 1.0804, 0.9214, 1.0767, 0.9253, 1.0727, 0.929, 1.0682]  # 定义 theta 数组
    K = len(theta)  # 计算 theta 数组的长度
    result = 0  # 初始化结果
    for k in range(K):
        result += theta[k] * (k + 2)*((-1)**k / (2*math.factorial(k)*math.factorial(k)*comb(2*k, k))) * (lambda_1 ** k) * (2 - lambda_1)** k # 计算 h(λ)
    return result
# 绘制 h(λ) 函数的图像
lambda_values = np.linspace(0, 2, 100)  # 定义 λ 的取值范围和步长
h_values = [h(lambda_1) for lambda_1 in lambda_values]  # 计算 h(λ) 的取值
plt.plot(lambda_values, h_values, lw = 2.5)  # 绘制图像
plt.xlabel(r'$\lambda$',fontsize = 20)  # 设置 x 轴标签
plt.ylabel(r'$h(\lambda)$',fontsize = 20)  # 设置 y 轴标签
# plt.legend(['cora'],loc=(0.75, 0.8),frameon=False,fontsize = 15)
plt.xticks(np.arange(0, 2.1, step=0.5),fontsize = 15)
plt.yticks(fontsize=15) 
# plt.savefig('cora.pdf', dpi=1200, bbox_inches = 'tight')
plt.show()  # 显示图像
```
